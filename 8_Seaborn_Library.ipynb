## 1- Categorical Data

# 1-First Import Libraries 

```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
```

# 2-Load Tips Datasets

```python
tips =sns.load_dataset('tips')
tips.head()
```

# 3- Data Info

```python
tips.info()
```

# 4-Make BarPlot 
*   Mean Is Default ==> sns.barplot



```python
sns.barplot(x = 'sex',y ='total_bill',data = tips)
```

# 5.Change Mean from data

```python
sns.barplot(x = 'sex',y ='total_bill',data = tips,estimator=np.std)
```

# 6-Countplot

```python
sns.countplot(x = 'sex' ,data = tips)
```

# - To Know Columns Names Of Tips Data

```python
tips.columns
```

# -Using BoxPlot

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

# -Using BoxPlot But Change Colors

```python
sns.boxplot(x='day',y='total_bill',data=tips,palette='rainbow')
```

# -Using BoxPlot But Change Orientation ='H'

```python
sns.boxplot(data=tips,palette='rainbow',orient='h')
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker',palette='rainbow')
```

## 2- Distrbution Plots Data

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],kde=False)
```

```python
sns.distplot(tips['total_bill'],kde=False,bins =50)
```

# Joinplot

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
```

```python
#reg
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
#resid
sns.jointplot(x='total_bill',y='tip',data=tips,kind='resid')
```

## All Data Using PairPlot

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex')
```

```python
sns.pairplot(tips,hue='smoker')
```
