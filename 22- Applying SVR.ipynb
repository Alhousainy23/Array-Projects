# 5. Data
1. Divided Data Two Types (Input Data (x_data) , Output Data (y_data))
2. Splitting Data Two (Train ,Test)

```python
x_data = data.iloc[:,:-1]
y_data = data.iloc[:,-1]
print('The Shape Of X Data Is ==> ',x_data.shape)#
print('The Shape Of Y Data Is ==> ',y_data.shape)
```

# 6. Applying Linear Regression

```python
from sklearn.linear_model import LinearRegression ,SGDRegressor
from sklearn.metrics import mean_absolute_error , mean_squared_error , median_absolute_error , accuracy_score , r2_score
x1_train,x1_test ,y1_train,y1_test = train_test_split(x_data ,y_data ,test_size=0.25 ,random_state=0)
LinearregressionModel = LinearRegression()
LinearregressionModel.fit(x1_train,y1_train)
#==============================================================================================================
print('Linear Regression Train Score is : ' , LinearregressionModel.score(x1_train, y1_train))
print('Linear Regression Test Score is : ' , LinearregressionModel.score(x1_test, y1_test))
print('Linear Regression Coef is : ' , LinearregressionModel.coef_)
print('Linear Regression intercept is : ' , LinearregressionModel.intercept_)
print('----------------------------------------------------')
#Calculating Prediction
y_pred = LinearregressionModel.predict(x1_test)
print('Real Value for Linear Regression is      : ' ,list(y1_test[:5]))
print('Predicted Value for Linear Regression is : ' ,list(y_pred[:5]))
print('----------------------------------------------------')
#----------------------------------------------------
#Calculating Mean Absolute Error
MAEValue = mean_absolute_error(y1_test, y_pred, multioutput='uniform_average') # it can be raw_values
print('Mean Absolute Error Value is : ', MAEValue)
print('----------------------------------------------------')
#----------------------------------------------------
#Calculating Mean Squared Error
MSEValue = mean_squared_error(y1_test, y_pred, multioutput='uniform_average') # it can be raw_values
print('Mean Squared Error Value is : ', MSEValue)
#----------------------------------------------------
print('----------------------------------------------------')
#Calculating Median Squared Error
MdSEValue = median_absolute_error(y1_test, y_pred)
print('Median Squared Error Value is : ', MdSEValue )
#-----------------------------------------------------------------------
print('----------------------------------------------------')
R2_Score = r2_score(y1_test,y_pred)
print('The Accuracy Of The Program Is ==> ',R2_Score*100)
```

# Applying SGD

```python
# The 'loss' parameter of SGDRegressor must be a str among {'squared_epsilon_insensitive', 'huber', 'squared_error', 'epsilon_insensitive'}
x_train,x_test ,y_train,y_test = train_test_split(x_data ,y_data ,test_size=0.25 ,random_state=0)
SGD = SGDRegressor(loss='huber',alpha=0.01,max_iter=300000,verbose=2,epsilon=0.001)
SGD.fit(x_train,y_train)
#Calculating Details
print('SGD Regression Train Score is : ' , SGD.score(x_train, y_train))
print('----------------------------------------------------')
print('SGD Regression Test Score is : ' , SGD.score(x_test, y_test))
print('----------------------------------------------------')
print('SGD Regression Coef is : ' , SGD.coef_)
print('----------------------------------------------------')
print('SGD Regression intercept is : ' , SGD.intercept_)
print('----------------------------------------------------')
#Calculating Prediction
y_pred = SGD.predict(x_test)
print('Predicted Value for SGD Regression is : ' , list(y_pred[:5]))
#----------------------------------------------------
#Calculating Mean Absolute Error
MAEValue = mean_absolute_error(y_test, y_pred, multioutput='uniform_average') # it can be raw_values
print('Mean Absolute Error Value is : ', MAEValue)
#----------------------------------------------------
print('----------------------------------------------------')
#Calculating Mean Squared Error
MSEValue = mean_squared_error(y_test, y_pred, multioutput='uniform_average') # it can be raw_values
print('Mean Squared Error Value is : ', MSEValue)
#----------------------------------------------------
print('----------------------------------------------------')
#Calculating Median Squared Error
MdSEValue = median_absolute_error(y_test, y_pred)
print('Median Squared Error Value is : ', MdSEValue )
print('----------------------------------------------------')
#=======================================================================================================
R2_Score = r2_score(y_test,y_pred)
print('The Accuracy Of The Program Is ==> ',R2_Score*100)
```

```python
from sklearn.neural_network import MLPRegressor
MLPModel = MLPRegressor(hidden_layer_sizes=(200,5),activation='identity',solver='lbfgs',tol=0.00001,verbose=1,random_state=None , max_iter=100000,alpha=0.001,learning_rate_init=0.002)
x2_train,x2_test ,y2_train,y2_test = train_test_split(x_data ,y_data ,test_size=0.25 ,random_state=0)
MLPModel.fit(x2_train,y2_train)
#Calculating Details
print('MLPRegressorModel Train Score is : ' , MLPModel.score(x2_train, y2_train))
print('MLPRegressorModel Test Score is : ' , MLPModel.score(x2_test, y2_test))
print('MLPRegressorModel loss is : ' , MLPModel.loss_)
print('MLPRegressorModel No. of iterations is : ' , MLPModel.n_iter_)
print('MLPRegressorModel No. of layers is : ' , MLPModel.n_layers_)
print('MLPRegressorModel last activation is : ' , MLPModel.out_activation_)
print('----------------------------------------------------')
#Calculating Prediction
y2_pred = MLPModel.predict(x2_test)
print('Predicted Value for MLPRegressorModel is : ' , y2_pred[:10])
R2_Score = r2_score(y2_test,y2_pred)
print('The Accuracy Of The Program Is ==> ',R2_Score*100)
```

```python
#The 'criterion' parameter of RandomForestRegressor must be a str among {'poisson', 'absolute_error', 'squared_error', 'friedman_mse'}. Got 'gini' instead.
from sklearn.ensemble import RandomForestRegressor
RF = RandomForestRegressor(n_estimators=10000,criterion='friedman_mse',max_depth=None,random_state=0,n_jobs=-1)
x3_train,x3_test ,y3_train,y3_test = train_test_split(x_data ,y_data ,test_size=0.25 ,random_state=0)
RF.fit(x3_train,y3_train)
#Calculating Details
print('RandomForestClassifierModel Train Score is : ' , RF.score(x3_train, y3_train))
print('RandomForestClassifierModel Test Score is : ' , RF.score(x3_test, y3_test))
print('RandomForestClassifierModel features importances are : ' , RF.feature_importances_)
print('----------------------------------------------------')
#Calculating Prediction
y_pred = RF.predict(x3_test)
print('Predicted Value for RandomForestClassifierModel is : ' , y_pred[:5])
print('----------------------------------------------------')
R2_Score = r2_score(y3_test,y_pred)
print('The Accuracy Of The Program Is ==> ',R2_Score*100)
#====================================================================================
```
