```python
#dropping the id and date column
data=dataset.drop(['id','date'],axis=1)
data
```

```python
#understanding the distribution with seaborn
with sns.plotting_context("notebook",font_scale=2.5):
    g = sns.pairplot(dataset[['sqft_lot','sqft_above','price','sqft_living','bedrooms']], 
                 hue='bedrooms', palette='tab20',size=6)
g.set(xticklabels=[]);
```

```python
# Make Scaling For Data
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
Scalar1 = StandardScaler()
data_new = Scalar1.fit_transform(data)
#Scalar2 = StandardScaler.fit_transform(data)
#separating independent and dependent variable
x = data_new[:,1:]
y = data_new[:,0]
print('The Shape Of X Data Is ==> ',x.shape)
print('The Shape Of Y Data Is ==> ',y.shape)
#splitting dataset into training and testing dataset
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=0)
```

```python
x_train = Scalar1.fit_transform(x_train)
x_test  = Scalar1.fit_transform(x_test)
#y_train = Scalar1.fit_transform(y_train)
#y_test  = Scalar1.fit_transform(y_test)
print('\t\t The X Trian Data Is :\n',x_train)
print('***'*30)
print('\t\t The X Test Data Is :\n',x_test)
print('***'*30)
print('The Y Trian Data Is ==> ',y_train)
print('***'*30)
print('The Y Test Data Is ==> ',y_test)
print('***'*30)
```

```python
#{'adam', 'sgd', 'lbfgs'}. Got 'lbfs' instead.
#{'relu', 'tanh', 'logistic', 'identity'}
from sklearn.neural_network import MLPRegressor
MLPRegressor_Model = MLPRegressor(shuffle=True,hidden_layer_sizes=(800,15),activation='relu',solver='lbfgs',alpha=0.03,batch_size='auto',max_iter=10000,random_state=0,tol=0.00005)
MLPRegressor_Model.fit(x_train,y_train)
```

```python
print('MLPRegressorModel Train Score is : ' , MLPRegressor_Model.score(x_train, y_train))
print('MLPRegressorModel Test Score is : ' , MLPRegressor_Model.score(x_test, y_test))
print('MLPRegressorModel loss is : ' , MLPRegressor_Model.loss_)
print('MLPRegressorModel No. of iterations is : ' , MLPRegressor_Model.n_iter_)
print('MLPRegressorModel No. of layers is : ' , MLPRegressor_Model.n_layers_)
print('MLPRegressorModel last activation is : ' , MLPRegressor_Model.out_activation_)
```

```python
from sklearn.metrics import confusion_matrix
#Calculating Prediction
from sklearn.metrics import r2_score
y_pred1 = MLPRegressor_Model.predict(x_test)
print('The Real Value Before Applying The Model Is ==>  ',list(y_test[:5]))
print('Predicted Value for Ridge Regression is ==> ' , list(y_pred1[:5]))
print('***'*30)
R_Score = r2_score(y_test,y_pred1)
print('The Accuracy Of The Program Is ==> ',R_Score)
```
