# Label Encoder

```python
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
data['diagnosis'] = LE.fit_transform(data['diagnosis'])
```

# Divided Data

```python
x = data.iloc[:,1:]
y = data.iloc[:,0]
#from sklearn.preprocessing import StandardScaler
#Scalar1 = StandardScaler()
#x = Scalar1.fit_transform(x)
```

# Splitting My Data 

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,shuffle=True,test_size=0.1,random_state=0)
```

```python
model = tf.keras.models.Sequential()
#loss = BinaryCrossentropy
#kernal_int = 'random_uniform
#model.add(tf.keras.layers.Dense(256,kernel_initializer='random_normal',kernel_regularizer=regularizers.L1(0.01),bias_initializer=tf.keras.initializers.Ones(),activation=tf.nn.relu))# Layer 1
model.add(tf.keras.layers.Dense(256,kernel_initializer='random_normal',kernel_regularizer=regularizers.L1(0.001),bias_initializer=tf.keras.initializers.Ones(),activation=tf.nn.relu))# Layer 1
model.add(tf.keras.layers.Dense(128,kernel_initializer='random_normal',kernel_regularizer=regularizers.L1(0.001),bias_initializer=tf.keras.initializers.Ones(),activation=tf.nn.relu))# Layer 2
model.add(tf.keras.layers.Dense(64,kernel_initializer='random_normal',kernel_regularizer=regularizers.L1(0.001),bias_initializer=tf.keras.initializers.Ones(),activation=tf.nn.relu))# Layer 3
model.add(tf.keras.layers.Dense(32,kernel_initializer='random_normal',kernel_regularizer=regularizers.L1(0.001),bias_initializer=tf.keras.initializers.Ones(),activation=tf.nn.relu))# Layer 4
model.add(tf.keras.layers.Dense(16,kernel_initializer='random_normal',kernel_regularizer=regularizers.L1(0.001),bias_initializer=tf.keras.initializers.Ones(),activation=tf.nn.relu))# Layer 5
model.add(tf.keras.layers.Dense(2,kernel_initializer='random_normal',kernel_regularizer=regularizers.l2(0.001),bias_initializer=tf.keras.initializers.Ones(),activation=tf.nn.sigmoid))# Layer 6
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0008),loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=100)
```

```python
evluate_loss,evaluate_accuracy=model.evaluate(x_test,y_test)
print('The Evaluate Loss Data Is ==> ',evluate_loss)
print('The Evalute Accuracy Data Is ==> ',evaluate_accuracy*100)
```

```python
#Calculating Prediction
y_pred = model.predict(x_test)
print('Real Value for LogisticRegressionModel is      : ' ,list(y_test[:10]))
print('Predicted Value for LogisticRegressionModel is : ' , y_pred[:3])
#================================================================================
```
