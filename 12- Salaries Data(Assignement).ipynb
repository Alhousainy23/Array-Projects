# Step 3.6 
1. How many Job Titles were represented by only one person in 2013? (e.g. Job Titles with only one occurence in 2013?)
* First Method 
* Second Method

```python
#First Method
data2 = data[data['Year'] == 2013]
jobs_group =data2.groupby('JobTitle').count()
jobs_count =jobs_group[jobs_group['Id'] == 1]
print('The Number Of Jobs Were Represented by only one person in 2013 ==> ',jobs_count.count()['Id']) 
#========================================================================================================
print('***'*20)
#Second Method
print('The Number Of Jobs Were Represented by only one person in 2013 ==> ',sum(data[data['Year']==2013]['JobTitle'].value_counts() == 1) )
#data3=data[data['Year'] == 2013]['JobTitle']
#print(data[data['Year']==2013]['JobTitle'].value_counts().sum())
```
