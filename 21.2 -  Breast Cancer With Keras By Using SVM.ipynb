```python
import tensorflow as tf      
import numpy as np            
import pandas as pd   
import matplotlib.pyplot as plt
import seaborn as sns 
from sklearn.model_selection import train_test_split 
from keras import regularizers
```

```python
data = pd.read_csv('D:\\Programming\\Machine Learning Projects\\My Projects\\10- CNN (Tensorflow - Keras)\\Data\\breast-cancer.csv')
data.head()
```

```python
data.drop('id', axis=1, inplace=True)
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
data['diagnosis'] = LE.fit_transform(data['diagnosis'])
data
```

```python
x = data.iloc[:,1:]
y = data.iloc[:,0]
x_train,x_test,y_train,y_test = train_test_split(x,y,shuffle=True,test_size=0.20,random_state=0)
```

# Create Neural Network 

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(512,kernel_regularizer=regularizers.l2(0.1),activation=tf.nn.relu),
    tf.keras.layers.Dense(256,kernel_regularizer=regularizers.l2(0.5),activation=tf.nn.relu),
    tf.keras.layers.Dense(256,kernel_regularizer=regularizers.l2(0.2),activation=tf.nn.relu),
    tf.keras.layers.Dense(128,kernel_regularizer=regularizers.l2(0.2),activation=tf.nn.relu),
    tf.keras.layers.Dense(128,kernel_regularizer=regularizers.l2(0.5),activation=tf.nn.relu),
    tf.keras.layers.Dense(64,kernel_regularizer=regularizers.l2(1.0),activation=tf.nn.relu),
    tf.keras.layers.Dense(32,kernel_regularizer=regularizers.l2(1.0),activation='linear'),
    tf.keras.layers.Dense(2,kernel_regularizer=regularizers.l2(1.0),activation='linear')]
)
model.compile(optimizer=tf.keras.optimizers.Adadelta(learning_rate=0.5),loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=150)
```

```python
evluate_loss,evaluate_accuracy=model.evaluate(x_test,y_test)
print('The Evaluate Loss Data Is ==> ',evluate_loss)
print('The Evalute Accuracy Data Is ==> ',evaluate_accuracy*100)
```

```python
#Calculating Prediction
y_pred = model.predict(x_test)
print('Real Value for LogisticRegressionModel is      : ' ,list(y_test[:10]))
print('Predicted Value for LogisticRegressionModel is : ' , y_pred[:3])
#================================================================================
```
