```python
import matplotlib.pyplot as plt    
import numpy as np            
x= np.array([1,5,8,7])
y = np.array([10,15,20,17])
plt.style.use('ggplot')
ax=plt.axes()
ax.set(xlabel='X Axsis',ylabel='Y A xsis',title = 'Example 1')
plt.plot(x,y , marker = '*',ms=20)
```

```python
x= np.array([1,5,8,7])
#y = np.array([10,15,20,17])
x1 = np.array([2,1,4,5,8,7,9])
#y1= np.array(5,10,15,20,25,30,35)
plt.style.use('ggplot')
ax=plt.axes()
ax.set(xlabel='X Axsis',ylabel='Y A xsis',title = 'Example 1')
plt.plot(x, marker = '*',ms=20, color = 'red',label='Line 1')
plt.plot(x1,marker='^',ms=10,  color= 'green',label = 'Line 2')
ax.legend()
```
